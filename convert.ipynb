```python
import pandas as pd
birdstrikes = pd.read_csv(r'data/birdstrikes_Final.csv')
```

```python
birdstrikes = birdstrikes[["Airport Name","Flight Date","Origin State"]]
```

```python
birdstrikes['Airport Name'] = birdstrikes['Airport Name'].str.replace(' ARPT', ' AIRPORT')
birdstrikes['Airport Name'] = birdstrikes['Airport Name'].str.replace(' INTL', ' INTERNATIONAL')
```

```python
birdstrikes['Flight Date'] = pd.to_datetime(birdstrikes['Flight Date'])
birdstrikes['Flight Date'].astype('datetime64[ns]')
```

```python
birdstrikes["year"] = birdstrikes["Flight Date"].dt.year
birdstrikes
```
